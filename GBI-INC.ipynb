# Mouse rat

```python
link="http://gbi-inc.com/products/polymer-hrp-and-ap-detection-kits/for-mouse-rat-or-other-tissue/for-chicken-antibody"
source=requests.get(link).text
soup=BeautifulSoup(source,'lxml')
blurb=soup.findAll('div',{'class':'blurb'})
```

```python
def updated_header(soup):
    head_data=[]
    headline=soup.findAll('div',{"class":'headline subsub'})
    for head in headline:
        dead_dict={'Used For':head.text}
        head_data.append(dead_dict)
    return head_data    
```

```python
def updated_feat_protocol(soup):
    feat=[]
    col_p=[]
    blurb=soup.findAll('div',{'class':'blurb'})
    for blur in blurb:

        prot={}
        features={}
        unorder=blur.find('ul')

        order=blur.find('ol')

        unord_list=[]
        order_list=[]
        if unorder is not None:
            unorder=unorder.findAll('li')
            for ul in unorder:
                unord_list.append(ul.text)
            unord_list='|'.join(unord_list)
            features.update({'Features':unord_list})
        if order is not None: 
            order=order.findAll('li')
            for ol in order:
                order_list.append(ol.text)
            order_list='|'.join(order_list)
            prot.update({'Protocols':order_list})

        col_p.append(prot)
        feat.append(features)

    return feat,col_p
    
#return features,prot
```

```python

link="http://gbi-inc.com/products/polymer-hrp-and-ap-detection-kits/for-mouse-rat-or-other-tissue"
source=requests.get(link).text
soup=BeautifulSoup(source,'lxml')
names=soup.find('ul',{"class":"pageList level3"})
names=names.findAll('a')
name_data_mouserat=[]
for name in names:
    name="http://gbi-inc.com"+name['href']
    name_data_mouserat.append(name)
    
name_data_mouserat    
```

```python
all_data=[]
for num,n_data in enumerate(name_data_mouserat):
    #n_data="http://gbi-inc.com/products/polymer-hrp-and-ap-detection-kits/for-mouse-rat-or-other-tissue/for-mouse-antibody"
    source=requests.get(n_data).text
    soup=BeautifulSoup(source,'lxml')

    products=soup.findAll("div",{"class":'products'})
    for number_p,pro_num in enumerate(products):

        pro_num=pro_num.find("table")
        tab_head=pro_num.find('tr',{"class":"header"})
        tab_head=tab_head.findAll('td')
        table_head=[]
        for t_h in tab_head[:-1]:
            table_head.append(t_h.text)
        pro_num=pro_num.findAll('tr')  
        pro_num=pro_num[1:]
        pr_dat=[]
        for pro in pro_num:
            price_tag=[]
            per_data={}
            td=pro.findAll('td')
            for j,t in enumerate(td):
                per_data.update({"URL":n_data})

                price_list=t.findAll('a',{"class":'btn'})
                #print(price_list)
                if len(price_list)>0:
                    pric=[]
                    for cash,prz in enumerate(price_list):
                        pric.append(prz['title'])
                    price_tag.append(pric)    
                    #print(pric)    
                else:
                    pass
                alpha=t.find('a')
                if alpha is not None and alpha.has_attr('href'):


                    per_data.update({"Data Sheet":"http://gbi-inc.com"+alpha['href']})
                elif alpha is not None and alpha.has_attr('data-pic'):

                    img_url=alpha['data-pic']
                    img_url=img_url.replace(" ","%20")
                    img_url="http://gbi-inc.com/images/products/"+img_url
                    per_data.update({"Image":img_url})
                try:

                    per_data.update({table_head[j]:t.text})

                except:
                    pass

            #for le in range(len(main)):
            head=updated_header(soup)
            per_data.update(head[number_p])
            feat_p,col_p=updated_feat_protocol(soup)
            if len(feat_p)>0:
                
                per_data.update(feat_p[number_p])
            if len(col_p)>0:
                
                per_data.update(col_p[number_p])    
            
            for product_price_list in price_tag:

                for pl in product_price_list:
                    price_and_other={}
                    price_and_other.update({"Price_":pl})
                    price_and_other.update(per_data)
                    all_data.append(price_and_other)
    print("process no "+str(num)+" completed")    
    
        



#price_tag
```
