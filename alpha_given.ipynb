```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import csv
import numpy as np
import os
```

```python
data=pd.read_excel("ITK_Alpha_newproducts.xlsx")
article_number=list(data["Article number"])
filename='alpha_info.txt'
```

```python
if not os.path.isfile(filename):
    with open(filename, 'a') as file:
        file.write("0,FERT16-S \n")
        print("\n")
        start=0
else:
    with open(filename, 'r') as f:
        lines = f.read().splitlines()
        start = lines[-1]
        start=start.split(',')[0]
```

```python


def data_scraping():
    
    final=[]
    for x in range(int(start),len(article_number)+1):
        if x == start:
            
            continue
        else:
            with open(filename, 'a') as file:
                
                file.write(str(x)+","+str(article_number[x])+"\n")
                print("\n")
       #scraping table 
        try:

            s="https://www.4adi.com/render.cz?method=searchProducts&categoryId=&keywords="+article_number[x]
            source=requests.get(s).text
            soup=BeautifulSoup(source,'lxml')
            #soup
            table_rows=soup.findAll("tr",{"class":"item"})
            info=[]
            links=[]
            u=[]
            for i in table_rows:
                row=[]
                alpha=[]
                atag=i.findAll('a')
                for a in atag:

                    alpha.append(a['href'])
                links.append(alpha[0])
                td=i.findAll('td')
                for t in td:

                    row.append(t.text)
                row.append(alpha[1])    
                info.append(row)    

    #cleaning table text as it will be very messy
            for i in info:
                 for d,j in enumerate(i):
                        if '\t' in j or "\n" in j:
                            j=j.split()
                            j=' '.join(j)
                            i[d]=j

    #extracting information of each product                       

            for i in range(len(links)):
                ss="https://www.4adi.com/4adi/"+links[i]
                source=requests.get(ss).text
                soup=BeautifulSoup(source,'lxml')
                table=soup.findAll("table")
                table=table[1]
                rows=table.findAll("tr")
                for r in rows:

                    h=[]
                    td_m=r.findAll('td')
                    for td in td_m:
                        
                        info[i].append(td.text)


            for q in info:
                

                final.append(q)
            print("file number # "+str(x)+" completed")    
        except:
            
            continue

    return (final)                

            
            
    
           
            

                
```

```python
web=data_scraping()
print(web)
```

```python
listing=[]
for i,j in enumerate(web):
    v=[]
    #print(j)
    c_d=[]
    catalog_number=j[0]
    name=j[1]
    typee=j[2]
    size=j[4]
    price=j[5]
    pdf=j[6]
    v.append(j[7:])
    for x in v:
        dict1={}
        for g,n in enumerate(x):
            #print(g,n)
            try:
                g=g*2
                data=x[g]
                desc=x[g+1]
                dict1.update({data:desc})
            except:
                pass
        dict1.update({"Catalog Number":catalog_number,"Name":name,"Product Type":typee
                     ,"Size":size,"Price":price,"PDF REFERENCE":pdf}) 
        
    listing.append(dict1)
    
        
        
        
        
        
listing            
```

```python
pd.max_columns=150
human_data_output=pd.DataFrame(listing)
#human_data_output.iloc[]
human_data_output
```
