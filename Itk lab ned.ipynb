```python

def images(source):
    image_data={}
    soup=BeautifulSoup(source,'lxml')
    images=soup.findAll('img')
    if len(images)>3:
        images=images[3:]
        for x in range(len(images)):
            image_url="https://labned.com"+x['src']
            image_data.update({"Image_"+str(x+1):image_url})
    else:
        image_data.update({"Image_1":"No Image"})
    return image_data    
    
    
```

```python
def url(source):
    
    js=json.loads(source)
    fetch=js['result']['_result'][0]
    name=fetch['name']
    artno=fetch['artno']
    artno=artno.lower()
    name=name.lower()
    if '.' in name:
        name=name.replace('.','')
    name=name.split()
    name='-'.join(name)
    combine=name+"-"+artno
    url="https://labned.com/"+combine
    return url
    
```

```python
#link="https://labned.com/egfl6-human-rec-protein-ln4105361"
#source=requests.get(link).text
def pdf_sheet(source,num):
    soup=BeautifulSoup(source,'lxml')
    span=soup.findAll('span')
    pdf={}
    for s in span:
        if "PDF Datasheet" in s.text:
            sheet=s.text
            pdf_link="https://api.labned.com/img/pdf/41/"+num+".pdf"
            pdf.update({sheet:pdf_link})
    return pdf        
        

    
```
