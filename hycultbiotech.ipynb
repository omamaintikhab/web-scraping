```python
def pro_links(soup):
    links={}
    try:
        pdf=soup.findAll('div',{"class":"item-link"})
        pd=pdf[0].find('a')['href']
        try:
            links.update({"CoA-TDS":pd})
        except:
            pass
        for d in pdf[1:]:
            a=d.find('a')
            atext=a.text.replace("\n","")

        
            links.update({atext:a['href']})
    except:
        pass
    return links
pro_links(soup)
```

```python
#soup.findAll('table')
#application
#additional
def pro_tables(soup):
    tablez=soup.findAll('table')
    tab_data={}
    for table in tablez:
        try:
            #table=soup.find('table',{"class":"TFtable"})
            table=table.findAll('tr')
            for tab in table:
                th=tab.find('th').text
                th=th.split()
                th=' '.join(th)
                td=tab.find('td').text
                tab_data.update({th:td})
        except:
            pass
    return tab_data
pro_tables(soup)
```

```python
def pro_tit(soup):
    tit={}
    try:
        title=soup.find('h1',{"class":"page-title"}).text
        tit.update({"Title":title})
    except:
        pass
    return tit
pro_tit(soup)
```

```python
def pro_overview(soup):
    tit={}
    try:
        title=soup.find('div',{"class":"product attribute overview"}).text
        tit.update({"Overview":title})
    except:
        pass
    return tit
pro_overview(soup)
```

```python
def pro_quan(soup):
    quan={}
    try:
        quantity=soup.find('div',{"class":"product-quantity-text"}).text
        quantity=quantity.replace("\n",'')
        quantity=quantity.split(":")
        quan.update({quantity[0]:quantity[1]})
    except:
        pass
    return quan

```

```python
def pro_cat(soup):
    cat={}
    try:
        catalogue=soup.find('div',{"class":"product attribute sku"}).text
        catalogue=catalogue.replace("\n",'')
        catalogue=catalogue.split(":")
        cat.update({catalogue[0]:catalogue[1]})
    except:
        pass
    return cat
```
