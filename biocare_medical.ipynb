```python
def table_info(sou):
    
    source=requests.get(sou).text
    soup_level1=BeautifulSoup(source,'lxml')  
    table=soup_level1.findAll('table')
    print(len(table))
    table_data={}
    links={}
    international={}
    if len(table)==2:
        table_1=table[0]
        table_rows=table_1.findAll('tr')
        for tr in table_rows:
            
            th=tr.find('th')

            td=tr.find('td')
            table_data.update({th.text:td.text})
        table_2=table[1]
        table_rows2=table_2.findAll('tr')
        for tr2 in table_rows2:
            
            td2=tr2.find('td')
            a=tr2.find('a')
            links.update({td2.text:a['href']})


    if len(table)==1:
        

        table_2=table[0]
        table_rows2=table_2.findAll('tr')
        for tr2 in table_rows2:
            

            td2=tr2.find('td')
            a=tr2.find('a')
            links.update({td2.text:a['href']})

    if len(table)==3:
        

        table_1=table[0]
        table_rows=table_1.findAll('tr')
        for tr in table_rows:

            th=tr.find('th')

            td=tr.find('td')
            table_data.update({th.text:td.text})
        table_2=table[1]
        table_rows2=table_2.findAll('tr')
        for tr2 in table_rows2:

            td2=tr2.find('td')
            a=tr2.find('a')
            links.update({td2.text:a['href']})
        table_3=table[2]
        table_rows3=table_3.findAll('tr')
        for i in table_rows3:

            text=i.find("td")
            a_t=i.find('a')
            #print(a_t['href'])
            if a_t is not None or text is not None:
                
                international.update({text.text:a_t['href']})

    return international,table_data,links
```
