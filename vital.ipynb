```python
def price(source):
#l="https://www.virostat-inc.com/products1/clostridia-sp"
    #l="https://www.virostat-inc.com/products-1/mycobacteria-lam"
    #source=requests.get(l).text
    soup=BeautifulSoup(source,'lxml')
    a=soup.find("div",{"id":"productDetails"})
    a=a["data-variants"]
    a=a.replace('[','')
    a=a.replace(']','')

    a=a.split('}')
    cl=[]
    for i in a:
        if 'priceMoney' in i:
            j=i.split('value:')
            #i=i.replace('{',"")
            #j=i.split('value')
            for x in j:
                x=x.split('value')
                cl.append(x[1])
    price=[]
    for c in cl:
        c=c.replace(":",'')
        c=c.replace('""','')
        c=c.replace('"','')
        price.append(c)
    return price    
        


```

```python
#l="https://www.virostat-inc.com/products-1/mycobacteria-lam"
#source=requests.get(l).text
def title_name(source):
    title_product={}
    soup=BeautifulSoup(source,'lxml')
    title=soup.find('h1',{'class':"product-title"}).text
    title_product.update({"Title Name":title})
    return title_product
```

```python
def product_excerpt(source):
    soup=BeautifulSoup(source,'lxml')
    s=soup.find('div',{'class':'product-excerpt'})
    #type(s)
    p=str(s.find('p'))
    p=p.replace('<p class="" style="white-space:pre-wrap;">','')
    p=p.split('<a')
    p=p[0]
    p=p.split('<br/>')
    data={}
    for i in p[:-1]:
        if ':' in i:
            i=i.split(':')
            data.update({i[0]:i[1]})
        else:
            data.update({"Clonality":i})
    return data        
```

```python
def specification_link(source):
    soup=BeautifulSoup(source,'lxml')
    s=soup.find('div',{'class':'product-excerpt'})
    s=s.findAll('a')
    specification={}
    for i in s:
        j="https://www.virostat-inc.com"+i['href']
        specification.update({i.text:j})

    return specification
```

```python
def getPDFContent(path):
    content = ""
    num_pages = 1
    p = open(path, "rb")
    pdf = PyPDF2.PdfFileReader(p)
    for i in range(0, num_pages):
        content += pdf.getPage(i).extractText() + "\n"
    content = " ".join(content.replace(u"\xa0", " ").strip().split())     
    return content 
```

```python
# Path of the pdf 
PDF_file = "recpdf.pdf"
  
  
# Store all the pages of the PDF in a variable 
pages = convert_from_path(PDF_file, 500,first_page=True) 
  
# Counter to store images of each page of PDF to image 
image_counter = 1
  
# Iterate through all the pages stored above 
for page in pages:
    # Declaring filename for each page of PDF as JPG 
    # For each page, filename will be: 
    # PDF page 1 -> page_1.jpg 
    # PDF page 2 -> page_2.jpg 
    # PDF page 3 -> page_3.jpg 
    # .... 
    # PDF page n -> page_n.jpg 
    filename = "page_"+str(image_counter)+".jpg"
      
    # Save the image of the page in system 
    page.save(filename, 'JPEG') 
  
    # Increment the counter to update filename 
    image_counter = image_counter + 1
filelimit = image_counter-1
outfile = "out_text.txt"
f = open(outfile, "a")
for i in range(1, filelimit + 1):
    filename = "page_"+str(i)+".jpg"
    text = str(((pytesseract.image_to_string(Image.open(filename))))) 
    text = text.replace('-\n', '')
    f.write(text)
f.close()    
  
    
  
```

```python

  
import io
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage
import os
import sys, getopt

#converts pdf, returns its text content as a string
def convert(fname, pages=None):
    if not pages:
        pagenums = set()
    else:
        pagenums = set(pages)

    output = StringIO()
    manager = PDFResourceManager()
    converter = TextConverter(manager, output, laparams=LAParams())
    interpreter = PDFPageInterpreter(manager, converter)

    infile = file(fname, 'rb')
    for page in PDFPage.get_pages(infile, pagenums):
        interpreter.process_page(page)
    infile.close()
    converter.close()
    text = output.getvalue()
    output.close
    return text 
   
def convertMultiple(pdfDir, txtDir):
    if pdfDir == "": pdfDir = os.getcwd() + "\\" #if no pdfDir passed in 
    for pdf in os.listdir(pdfDir): #iterate through pdfs in pdf directory
        fileExtension = pdf.split(".")[-1]
        if fileExtension == "pdf":
            pdfFilename = pdfDir + pdf 
            text = convert(pdfFilename) #get string of text content of pdf
            textFilename = txtDir + pdf + ".docx"
            textFile = open(textFilename, "w") #make text file
            textFile.write(text) #write text to text file
			#textFile.close

pdfDir = "C:/Users/omama/Desktop/trypdf/pdf"
txtDir = "C:/Users/omama/Desktop/trypdf/ttxt"
convertMultiple(pdfDir, txtDir)
```

```python
def extract_text(self):
    """ Patched extractText() from pyPdf to put spaces between different text snippets.
    """
    text = u""
    content = self["/Contents"].getObject()
    if not isinstance(content, ContentStream):
        content = ContentStream(content, self.pdf)
    # Note: we check all strings are TextStringObjects.  ByteStringObjects
    # are strings where the byte->string encoding was unknown, so adding
    # them to the text here would be gibberish.
    for operands, operator in content.operations:
        if operator == "Tj":
            _text = operands[0]
            if isinstance(_text, TextStringObject):
                text += _text
        elif operator == "T*":
            text += "\n"
        elif operator == "'":
            text += "\n"
            _text = operands[0]
            if isinstance(_text, TextStringObject):
                text += operands[0]
        elif operator == '"':
            _text = operands[2]
            if isinstance(_text, TextStringObject):
                text += "\n"
                text += _text
        elif operator == "TJ":
            for i in operands[0]:
                if isinstance(i, TextStringObject):
                    text += i

        if text and not text.endswith(" "):
            text += " "  # Don't let words concatenate

    return text
```

```python
from PyPDF2 import PdfFileReader, PdfFileWriter

def merge_pdfs(paths, output):
    pdf_writer = PdfFileWriter()

    for path in paths:
        pdf_reader = PdfFileReader(path)
        for page in range(pdf_reader.getNumPages()):
            # Add each page to the writer object
            pdf_writer.addPage(pdf_reader.getPage(page))

    # Write out the merged PDF
    with open(output, 'wb') as out:
        pdf_writer.write(out)

if __name__ == '__main__':
    

    #paths = ['document1.pdf', 'document2.pdf']
    merge_pdfs(file_list, output='merged.pdf')
```
