```python
def kitscomponents(soup):
    kis={}
    try:
        kitss=[]
        kitcomponent=soup.find('table',{"id":"kit-component-table"})
        kitcomponent=kitcomponent.findAll('tr')
        for tr in kitcomponent[1:]:
            tdlist=[]
            kittd=tr.findAll('td')
            for td in kittd:
                td_text=td.text.split()
                td_text=' '.join(td_text)
                #print(td_text)
                tdlist.append(td_text)
            tdlist=' , '.join(tdlist)
            kitss.append(tdlist)
        kitss=' | '.join(kitss)    
        kis.update({"Kits Components":kitss})
    except:
        pass
    return kis


def tables(soup):
    properties={}
    try:
        tables=soup.findAll('table',{"id":"overview"})
        for tab in tables:
            ttable_rows=tab.findAll('tr')
            for tr in ttable_rows:

                table_td=tr.findAll('td')
                if len(table_td)==2:
                    t1=table_td[0].text
                    t1=t1.split()
                    t1=' '.join(t1)
                    t2=table_td[1].text
                    t2=t2.split()
                    t2=' '.join(t2)

                    properties.update({t1:t2})
    except:
        pass
    return properties

def product_images(soup):
    image_data={}
    try:

        images=soup.findAll('div',{"class":"section-image-container"})
        for n,img in enumerate(images):
            paragraphs=[]
            paras=img.findAll('p')
            for p in paras:
                paragraphs.append(p.text)
            paragraphs=','.join(paragraphs)
            imgsrc=img.find('a')["href"]
            image_data.update({"Image_"+str(n+1):imgsrc,"Image_"+str(n+1)+"_description":paragraphs})
    except:
        pass
    return image_data 

def product_citation(soup):
    pro_citations={}
    try:
        citate=[]
        citations=soup.findAll('div',{"class":"citations"})
        for cit in citations:
            cit_text=cit.text.replace("\t","")
            cit_text=cit_text.replace("\n","")
            citate.append(cit_text)
        citate=' | '.join(citate)
        pro_citations.update({"Product Citation":citate})
    except:
        pass
    return pro_citations

def sizes(soup):
    finish=[]
    try:
        price=[]
        conjugate=[]
        size=[]
        sku=[]
        
            
        dis=soup.find('form',{"class":"variations_form cart"})["data-product_variations"].split(",")
        dis
        for d in dis:
            if '"display_price":' in d:
                prize=d.split(":")

                price.append(prize[-1])
            elif  '"attribute_pa_conjugate":' in d:
                conj=d.split(":")
                conj=conj[-1].replace("}","")
                conj=conj.replace('"',"")
                conjugate.append(conj)
            elif '"attribute_pa_size":' in d:
                sizing=d.split(":")
                sizing=sizing[-1].replace('"',"")
                sizing=sizing.replace('%c',"")
                sizing=sizing.replace('%b',"")
                size.append(sizing)
            elif '"sku":' in d:
                skuu=d.split(":")
                skuu=skuu[-1]
                sku.append(skuu)

        for i in range(len(price)):
            temp={}
            try:
                
                temp.update({"Sku":sku[i]})
                
            except:
                pass
            try:
                temp.update({"Conjugate":conjugate[i]})
            except:
                pass
            try:
                temp.update({"Price":price[i]})
            except:
                pass
            try:
                temp.update({"Size":size[i]})
            except:
                pass
                
                
            finish.append(temp)
    except:
        pass
    return finish    
```
