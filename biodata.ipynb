```python
def product_title(sou):
    title_name={}
    source=requests.get(sou).text
    soup_level1=BeautifulSoup(source,'lxml')  
    title=soup_level1.find('h1',{"class":"product-title entry-title"})
    if title is not None:
        title=title.text
        
        if '\n' in title:
            title=title.replace('\n',"")


                
        if '\t' in title:
            

            title=title.replace('\t',"")
        title_name.update({"Title":title})
        return title_name
    else:
        title_name.update({"Title":None})
        return title_name
```

```python
def table_information(sou):

    source=requests.get(sou).text
    soup_level1=BeautifulSoup(source,'lxml')  
    table=soup_level1.findAll('table')
    table_data={}
    for tab in table:
        
        table_rows=tab.findAll('tr')
        for tr in table_rows:
            
            th=tr.find('th')

            td=tr.find('td')
            if th is not None and td is not None:
                
                
                table_data.update({th.text:td.text})
            else:
                lnk=tr.find('a')
                if lnk is not None:
                    table_data.update({lnk.text:lnk['href']})
                    

    return table_data                            
                    
                    



            
           
           
                
                
            
            
                
            
           
    
        
          
```

```python
def Category(sou):
    source=requests.get(sou).text
    soup_level1=BeautifulSoup(source,'lxml')
    soup_level1
    span=soup_level1.findAll('span',{"class":["sku_wrapper"]})
    sku_cat={}
    for s in span:
        sku=s.findAll('span',{"class":"sku"})
        for i in sku:
            sku_cat={"SKU":i.text}     


    span_cat=soup_level1.findAll('span',{"class":"posted_in"})
    sp_list=[]
    sp_dict={}
    for a in span_cat:

        ast=a.findAll('a')
        for ar in ast:

            sp_list.append(ar.text)
        category=' '.join(sp_list)
        sp_dict={"Category":category}

    sp_dict.update(sku_cat)    
    return (sp_dict)    

    
```

```python
def Description_Reference(sou):
    desc_ref={}
    sou="https://biocare.net/product/nerve-growth-factor-receptor-antibody/"
    source=requests.get(sou).text
    soup_level1=BeautifulSoup(source,'lxml')
    para=soup_level1.findAll('p')
    desc=para[3].text
    desc_ref.update({"Description":desc})

    for i in para:
        if "1." in i.text:
            desc_ref.update({"Reference":i.text})

    return desc_ref
```

```python
def Product_Image(sou):
    image={}
    #sou="https://biocare.net/product/nerve-growth-factor-receptor-antibody/"
    source=requests.get(sou).text
    soup_level1=BeautifulSoup(source,'lxml')
    img=soup_level1.find('img',{"class":"wp-post-image"})
    if img is not None:
        image.update({'Image':img['src']})
    return image
    
    
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import csv
import numpy as np
import os
#Make sure that the address and directory of file is right
data=pd.read_excel("ITK_Alpha_newproducts.xlsx")
article_number=list(data["Article number"])
filename='alpha_info.txt'
if not os.path.isfile(filename):
    with open(filename, 'a') as file:
        file.write("0,FERT16-S \n")
        print("\n")
        start=0
else:
    with open(filename, 'r') as f:
        lines = f.read().splitlines()
        start = lines[-1]
        start=start.split(',')[0]


def data_scraping():
    final = []
    #making sure that the script starts from where  it ends previously
    #for x in range(int(start), len(article_number) + 1):
    for x in range(0,10):
        if x == start:

            continue
        else:
            with open(filename, 'a') as file:

                file.write(str(x) + "," + str(article_number[x]) + "\n")
                print("\n")
        # scraping table
        try:

            s = "https://www.4adi.com/render.cz?method=searchProducts&categoryId=&keywords=" + article_number[x]
            source = requests.get(s).text
            soup = BeautifulSoup(source, 'lxml')
            # soup
            table_rows = soup.findAll("tr", {"class": "item"})
            info = []
            links = []
            u = []
            for i in table_rows:
                row = []
                alpha = []
                atag = i.findAll('a')
                for a in atag:
                    alpha.append(a['href'])
                links.append(alpha[0])
                td = i.findAll('td')
                for t in td:
                    row.append(t.text)
                row.append(alpha[1])
                info.append(row)

                # cleaning table text as it will be very messy
            for i in info:
                for d, j in enumerate(i):
                    if '\t' in j or "\n" in j:
                        j = j.split()
                        j = ' '.join(j)
                        i[d] = j

            # extracting information of each product

            for i in range(len(links)):
                ss = "https://www.4adi.com/4adi/" + links[i]
                source = requests.get(ss).text
                soup = BeautifulSoup(source, 'lxml')
                table = soup.findAll("table")
                table = table[1]
                rows = table.findAll("tr")
                for r in rows:

                    h = []
                    td_m = r.findAll('td')
                    for td in td_m:
                        info[i].append(td.text)

            for q in info:
                final.append(q)
            print("file number # " + str(x) + " completed")
        except:

            continue

    return (final)



web=data_scraping()
#print(web)

listing = []
# creating data frame
for i, j in enumerate(web):
    v = []
    # print(j)
    c_d = []
    catalog_number = j[0]
    name = j[1]
    typee = j[2]
    size = j[4]
    price = j[5]
    pdf = j[6]
    v.append(j[7:])
    for x in v:
        dict1 = {}
        for g, n in enumerate(x):
            # print(g,n)
            try:
                g = g * 2
                data = x[g]
                desc = x[g + 1]
                dict1.update({data: desc})
            except:
                pass
        dict1.update({"Catalog Number": catalog_number, "Name": name, "Product Type": typee
                         , "Size": size, "Price": price, "PDF REFERENCE": pdf})

    listing.append(dict1)

pd.max_columns=150
alpha_dataframe=pd.DataFrame(listing)
print(alpha_dataframe)
if not os.path.isfile('alpha_csvfile.csv'):
    file = open('alpha_csvfile.csv','a+')
    alpha_dataframe.to_csv(file,header=True)
else:
    file = open('alpha_csvfile.csv', 'a+')
    alpha_dataframe.to_csv(file,header=False)





```
