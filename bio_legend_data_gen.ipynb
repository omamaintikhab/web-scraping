```python
def Product_Image(source):
    soup=BeautifulSoup(source,'lxml')
    image=soup.find("div",{"id":"productInfo"})
    image=image.findAll('img')
    image_data={}
    if image is not None:
        for i,img in enumerate(image): 
            image_data.update({"Image_"+str(i+1):img['src']})
    else:
        image_data.update({"Image_1":"No Image"})
    return image_data        
```

```python
def Related_Protocols(source):
    #source=requests.get(bio_legend_url[0]).text
    soup=BeautifulSoup(source,'lxml')

    rel_pro=soup.find('div',{"id":"productRelatedProtocols"})
    related_product_protocols={}
    if rel_pro is not None:
        related_protocol=rel_pro.findAll('a')
        related_protocol=related_protocol[1:]
        
        for i in related_protocol:
            text=i.text
            if '\n' in text or '\r' in text:
                text=text.split()
                text=' '.join(text)
                related_product_protocols.update({"Related Protocols":i['href']})
    else:
        related_product_protocols.update({"Related Protocols":"None"})
        

        #related_product_protocols.update({i.text:i['href']})
    return related_product_protocols    
```

```python
#source=requests.get(bio_legend_url[0]).text
#soup=BeautifulSoup(source,'lxml')
def Documentation(source):
    soup=BeautifulSoup(source,'lxml')
    prod_doc=soup.find('div',{"id":"productDocumentation"})
    alpha_=prod_doc.findAll('a')
    product_documentation={}
    if alpha_ is not None:
        for a in alpha_:
            
            product_documentation.update({a.text:a['href']})
            if a.text=='Safety Data Sheet':
                try:
                    url="https://www.biolegend.com"+a['href']
                    source=requests.get(url).text
                    soup=BeautifulSoup(source,'lxml')
                    safety_data_link=soup.findAll('a')
                    for sdl in safety_data_link:
                        if sdl.text=='United States (US) English (US)':

                            product_documentation.update({"Safety_pdf":sdl['href']})
                except:
                    pass
            
            
        return product_documentation    

    else:
        return product_documentation
        
```

```python
def Product_Details(source):
    soup=BeautifulSoup(source,'lxml')
    data_dict={}
    a=soup.find('div',{"id":"productDetails"})
    if a is not None:
        dd_list=[]
        dt_list=[]
        #a=soup.find('div',{"id":"antigenDetails"})
        dt=a.findAll('dt')
        dd=a.findAll('dd')
        for i in dt:
            dt_list.append(i.text)    


        for i in dd:
            para=i.findAll('p')
            if len(para)>0:
                para_list=[]
                for p in para:
                    para_list.append(p.text)
                    para_g="| ".join(para_list) 
                dd_list.append(para_g)



            elif len(para)<=0:

                list_items=i.findAll('li')
                if len(list_items)>0:
                    l=[]
                    for li in list_items:
                        l.append(li.text)
                    list_g=' |'.join(l)
                    dd_list.append(list_g)


                else:
                    dd_list.append(i.text)


        for j,i in enumerate(dd_list):




            if '\n' in i or '\r' in i:

                i=i.split()
                i=' '.join(i)
                dd_list[j]=i
        for j,i in enumerate(dt_list):

            if '\n' in i or '\r' in i:

                i=i.split()
                i=' '.join(i)
                dt_list[j]=i

        
        for i in range(len(dt)):
            data_dict.update({dt_list[i]:dd_list[i]})

        return data_dict
    else:
        return data_dict
      
```
