```python

for enum,i in enumerate(urls_list):
    lent=i[1]+1
    for pno in range(1,lent):
        
        src=i[0]+str(pno)
        source=requests.get(src).text
        soup=BeautifulSoup(source,'lxml')
        urls=soup.findAll('div',{"class":"u-lr"})
        for url in urls:
            product_url.append(url.find('a')["href"])
        print(pno)
    #print(enum+"completed")
    print("__________________________")

    
```

```python
def product_overviw(soup):
    para_data={}
    try:
        para=soup.find('div',{"id":"DESC"})
        para=para.findAll('p')

        for p in para:
            try:
                strong=p.find("strong").text
                strong=strong.lower()
                strong=strong.split()
                strong=' '.join(strong)
                
                #strong=strong.replace('" "',"")
                paras_text=p.text
                para_data.update({strong:paras_text})
            except:
                pass
            try:
                bold=p.find("b").text
                bold=bold.lower()
                bold=bold.split()
                bold=' '.join(bold)
                #bold=bold.replace('" "',"")
                
                paras_text1=p.text
                para_data.update({bold:paras_text1})
            except:
                pass
    except:
        pass
    return para_data

def product_guide(soup):
    guide={}
    try:
        pdfs=soup.find('div',{"id":"MANUAL"})
        pdfs=pdfs.findAll('a')
        for pdf in pdfs:
            guide.update({pdf.text:pdf["href"]})
    except:
        pass

    return guide   

def product_citation(soup):
    pro_cit={}
    try:
        temp_list=[]
        citations=soup.find('div',{"id":"EXTRA1"})
        citations=citations.findAll('p')
        for citation in citations:
            temp_list.append(citation.text)
        temp_list=' | '.join(temp_list)
        pro_cit.update({"Citations":temp_list})
    except:
        pass
    return pro_cit

def product_colonality(soup):
    left_data={}
    try:
        left_table=soup.find('div',{"class":"c-productinfo__info__desc--left"})
        left_table=left_table.find('table')
        left_table=left_table.findAll('tr')
        for tr in left_table:
            table_td=tr.findAll('td')
            left_data.update({table_td[0].text:table_td[1].text})
    except:
        pass
    return left_data   

def product_price_info(soup):
    row_data=[]
    headings=[]
    right_table=soup.find('div',{"class":"c-productinfo__info__desc--right"})
    right_table=right_table.find('table')
    right_table=right_table.findAll('tr')
    heading=right_table[0]
    heading=heading.findAll('td')
    for head in heading:
        headings.append(head.text)
    headings=headings[:3]
    right_table=right_table[1:]
    for tr in right_table:
        lis={}
        table_td=tr.findAll('td')
        for n,td in enumerate(table_td[:3]):
            lis.update({headings[n]:td.text})
        row_data.append(lis)
    return row_data


def product_images(soup):
    
    image_data={}
    try:
        images=soup.findAll('a',{"class":"tosrus"})
        images=images[1:]
        for n,image in enumerate(images):
            image_data.update({"Image_"+str(n):"https://www.epigentek.com/catalog/"+image['href'],"Image_"+str(n)+"_description":image["data-caption"]})
    except:
        pass
    return image_data  


def product_title(soup):
    pro_tit={}
    try:

        
        title=soup.find('h1',{"class":"c-productinfo__header__title"}).text
        title=title.split("\n")
        title=title[0]
        pro_tit.update({"Protduct Title":title})
    except:
        pass
    return pro_tit
        
    
```

```python
empty=[]
for num,urlz in enumerate(product_url):
    try:
        source=requests.get(urlz).text
        soup=BeautifulSoup(source,'lxml')
        prices=product_price_info(soup)
        title=product_title(soup)
        cit=product_citation(soup)
        left=product_colonality(soup)
        p_view=product_overviw(soup)
        pdfz=product_guide(soup)
        imgg=product_images(soup)
        for price in prices:
            price.update(title)
            price.update({"Product URL":urlz})
            if len(left)>0:
                price.update(left)
            if len(p_view)>0:
                price.update(p_view) 
            if len(cit)>0:
                price.update(cit)
            if len(pdfz)>0:
                price.update(pdfz)
            if len(imgg)>0:
                price.update(imgg)
            empty.append(price)
        print("Process number "+str(num)+" completed")
        

    except:
        print("___________________________________________")
        print("failed "+str(num))
        print(urlz)
        print("___________________________________________")
        continue
        
```
