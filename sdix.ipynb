```python
def tabular(soup):
    tab_data={}
    try:
        table=soup.find('table',{"class":"TFtable"})
        table=table.findAll('tr')
        for tab in table:
            th=tab.find('th').text
            th=th.split()
            th=' '.join(th)
            td=tab.find('td').text
            tab_data.update({th:td})
    except:
        pass
    
    return tab_data    
```

```python
def pro_desc(soup):
    descr={}
    try:
        desc=soup.find('div',{"class":"wpb_text_column wpb_content_element"})
        desc=desc.findAll('p')
        ptext=[]
        for p in desc:
            ptext.append(p.text)
        ptext=' | '.join(ptext)
        descr.update({"Description":ptext})
    except:
        pass
    if descr["Description"]=='':
        div=soup.findAll('div',{"class":"wpb_text_column wpb_content_element"})
        for d in div:
            if 'Alternative names:' in d.text:
                textt=d.text
                textt=textt.split("\n")
                textt=' | '.join(textt)
                descr["Description"]=textt
    return descr    
```

```python
def pdf(soup):
    links={}
    try:
        divs=soup.findAll("div",{"class":"vc_btn3-container vc_btn3-inline"})
        for d in divs:
            link=d.find('a')
            links.update({link.text:"https://www.sdix.com"+link['href']})
    except:
        pass
    return links    
    
```

```python
def pro_ref(soup):
    ref={}
    try:
        div=soup.findAll('div',{"class":"wpb_text_column wpb_content_element"})
        for d in div:
            if '“' in d.text:
                text=d.text
                text=text.split("”")
                text=' | '.join(text)
                text=text.replace("\n","")
                ref.update({"References":text})
    except:
        pass
    return ref    
```

```python
def pro_img(soup):
    pic={}
    try:
        img=soup.findAll('div',{"class":"wpb_single_image wpb_content_element vc_align_left"})
        for im in img:
            image=im.findAll('img')
            for num,i in enumerate(image):
                pic.update({"Image_"+str(num+1):i['src']})
    except:
        pass

    return pic        
```

```python
cal="https://www.sdix.com/portfolio/calibrator-1/"
source=requests.get(cal).text
soup=BeautifulSoup(source,'lxml')
```

```python
def sectable(soup):
    tab={}
    last=[]
    trs=soup.find('table').findAll('tr')
    for tr in trs:
        t1=tr.text
        if ':' in t1:
            splt=t1.replace("\n","")
            splt=splt.split(":")

            tab.update({splt[0]:splt[1]})
        else:
            tabtd=tr.findAll('td')
            tempt=[]
            for td in tabtd:
                tempt.append(td.text)
            tempt=' , '.join(tempt)
            last.append(tempt)
    last=' | '.join(last)
    tab['Analyte Info']=tab['Analyte Info']+" | "+last
    return tab            
            
        
```

```python
callibrator=['https://www.sdix.com/portfolio/calibrator-0/',
 'https://www.sdix.com/portfolio/calibrator-1/',
 'https://www.sdix.com/portfolio/calibrator-2/',
 'https://www.sdix.com/portfolio/calibrator-3/',
 'https://www.sdix.com/portfolio/calibrator-4/',
 'https://www.sdix.com/portfolio/calibrator-7/']    


```

```python
empty=[]
for num,x in enumerate(callibrator):
    kit={}
    source=requests.get(x).text
    soup=BeautifulSoup(source,'lxml')
    tablez=sectable(soup)
    data_s=pdf(soup)
    disc=pro_desc(soup)
    imgg=pro_img(soup)
    reff=pro_ref(soup)
    kit.update({"Product Url":x})
    if len(tablez)>0:
        kit.update(tablez)
    if len(disc)>0:
        kit.update(disc)
    if len(imgg)>0:
        kit.update(imgg)
    if len(data_s)>0:
        kit.update(data_s)
    if len(reff)>0:
        kit.update(reff)
    empty.append(kit)
    print("process no "+str(num)+" completed")
        
            
    
```
