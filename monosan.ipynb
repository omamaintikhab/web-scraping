```python
def specificity(soup):
    data_dict={}
    try:
        
        con_21_d=[]
        con_21_dtype=[]
        con_21=soup.find('div',{"id":"content-21"})
        con_21
        data=con_21.findAll('div',{"class":"attribute-table__data attribute-table__col"})
        for dta in data:
            con_21_d.append(dta.text)

        data_type=con_21.findAll('div',{"class":"attribute-table__label attribute-table__col"})    
        for dtat in data_type:
            con_21_dtype.append(dtat.text)

        data_d=[]
        data_type=[]
        for i in con_21_d:


            i=i.split()
            i=' '.join(i)
            data_d.append(i)
        for i in con_21_dtype:


            i=i.split()
            i=' '.join(i)
            data_type.append(i)    
            
        for i in range(len(data)):
            data_dict.update({data_type[i]:data_d[i]})
    except:
        pass
    return data_dict    
```

```python
def pdf_link(soup):
    data_sheet={}
    try:

        a=soup.find('a',{'class':'icon-pdf'})
        a=a['href']
        data_sheet.update({'Datasheet':a})
        
    except:
        data_sheet.update({'Datasheet':'None'})
    return data_sheet
```

```python
def storage(soup):
    data_dict22={}
    try:
        con_22_d=[]
        con_22_dtype=[]
        con_22=soup.find('div',{"id":"content-22"})
        con_22
        data=con_22.findAll('div',{"class":"attribute-table__data attribute-table__col"})
        for dta in data:
            con_22_d.append(dta.text)

        data_type=con_22.findAll('div',{"class":"attribute-table__label attribute-table__col"})    
        for dtat in data_type:
            con_22_dtype.append(dtat.text)


        data_d22=[]
        data_type22=[]
        for i in con_22_d:


            i=i.split()
            i=' '.join(i)
            data_d22.append(i)
        for i in con_22_dtype:


            i=i.split()
            i=' '.join(i)
            data_type22.append(i)    
            
        for i in range(len(data)):
            data_dict22.update({data_type22[i]:data_d22[i]})
    except:
        pass
    
        
    return data_dict22    
```

```python
def title_price(soup):
    price_title={}
    try:
        title=soup.find('h1',{'class':'page-title'})
        title=title.text
    except:
        title=''
    try:
        
        price=soup.find('span',{"class":'price'})
        price=price.text
    except:
        price=''
    price_title.update({'Title':title,'Price':price})
    return price_title
    
```
